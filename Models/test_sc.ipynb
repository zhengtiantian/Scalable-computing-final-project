```python
import numpy as np
import pandas as pd

correct_readings = np.random.uniform(1, 10, 100)
correct_classification = np.full(100, 1)
print(correct_readings)
print(correct_classification)

incorrect_low_readings = np.random.uniform(-10, 0, 50)
incorrect_low_classification = np.full(50, 0)
print(incorrect_low_readings)
print(incorrect_low_classification)

incorrect_high_readings = np.random.uniform(11, 100, 50)
incorrect_high_classification = np.full(50, 0)
print(incorrect_high_readings)
print(incorrect_high_classification)

all_readings = np.concatenate((correct_readings, incorrect_low_readings, incorrect_high_readings))
all_classification = np.concatenate((correct_classification, incorrect_low_classification, incorrect_high_classification))

print(all_readings)
print(all_classification)


df = pd.DataFrame(np.column_stack((all_readings, all_classification)))
print(df)

from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
# model = LogisticRegression(penalty='l2', C=0.1, solver='lbfgs')
# model = LinearSVC(C=1.0)
model = DecisionTreeClassifier()

X = df[0].values.reshape(-1, 1)

model.fit(X,df[1])

y_pred = model.predict(X)
df[3] = y_pred

from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,df[1]))

print(model.predict([[5]])[0])

import pickle
pkl_filename = "sc_proj_3_model.pkl"
with open(pkl_filename, 'wb') as file:
    pickle.dump(model, file)
```
